```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('sonar.csv')
df.head()
```

```python
x_input = df.iloc[:, :60].values
y_label = pd.get_dummies(df.iloc[:, 60])
```

```python
y_label
```

```python
model = Sequential()
model.add(Dense(units=60, input_dim=60, activation='relu'))
model.add(Dense(units=180, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=2, input_dim=60, activation='softmax'))
```

```python
print(x_input.shape, y_label.shape)
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
#train the model with fit():
model.fit(x_input, y_label, epochs=50)
```
